```python
non_scaled_df=data[non_scaled_features]
normalized_non_scaled=non_scaled_df / non_scaled_df.mean()
print(normalized_non_scaled.var())
```

```python
import sklearn
from sklearn.feature_selection import VarianceThreshold
# Create a VarianceThreshold feature selector
sel = VarianceThreshold(threshold=0.4)

# Fit the selector to normalized non_scaled_df
#scaled_df=data[scaled_features]
sel.fit(non_scaled_df/non_scaled_df.mean())

# Create a boolean mask
mask = sel.get_support()
# Apply the mask to create a reduced dataframe
reduced_df_var = non_scaled_df.loc[:, mask]

print("Dimensionality reduced from {} to {}.".format(non_scaled_df.shape[1], reduced_df_var.shape[1]))
```

```python
corr = reduced_df_var.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
# Draw the heatmap
fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corr, mask=mask,  cmap="coolwarm", center=0, linewidths=1, annot=True, fmt=".2f")
plt.show()
```

```python
# Calculate the correlation matrix and take the absolute value
corr_matrix = reduced_df_var.corr().abs()
# Create a True/False mask and apply it
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)
# List column names of highly correlated features (r > 0.90)
to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.90)]

# Drop the features in the to_drop list
reduced_df_var_cor = reduced_df_var.drop(to_drop, axis=1)
print("The reduced dataframe has {} columns.".format(reduced_df_var_cor.shape[1]))

reduced_df_var_cor
```
