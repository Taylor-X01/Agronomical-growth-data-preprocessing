```python
from scipy.cluster.vq import kmeans, vq
distortions = []
num_clusters = range(1, 7)

# Create a list of distortions from the kmeans function
for i in num_clusters:
    cluster_centers, distortion = kmeans(data[scaled_features],i)
    distortions.append(distortion)

# Create a DataFrame with two lists - num_clusters, distortions
elbow_plot = pd.DataFrame({'num_clusters': num_clusters, 'distortions': distortions})

# Creat a line plot of num_clusters and distortions
plt.rcParams.update({'font.size': 12})
plt.subplots(1,1,figsize=(20,15))
sns.lineplot(x='num_clusters', y='distortions', data = elbow_plot)
plt.xticks(num_clusters)
plt.ylabel('Distortions')
plt.xlabel('Number of clusters')
plt.savefig('elbow_plot.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Create centroids with kmeans for 2 clusters
cluster_centers,_ = kmeans(data[scaled_features], 2)

# Assign cluster labels and print cluster centers
data['cluster_labels'], _ = vq(data[scaled_features], cluster_centers)
print(data.groupby('cluster_labels')[scaled_features].mean())

# Plot cluster centers to visualize clusters
data.groupby('cluster_labels')[scaled_features].mean().plot(legend=True, kind='bar',figsize=(30, 25))
plt.savefig('Cluster_centers.png')
plt.show()
```

```python
non_scaled_df=data[non_scaled_features]
normalized_non_scaled=non_scaled_df / non_scaled_df.mean()
print(normalized_non_scaled.var())
```

```python
import sklearn
from sklearn.feature_selection import VarianceThreshold
# Create a VarianceThreshold feature selector
sel = VarianceThreshold(threshold=0.1)

# Fit the selector to normalized non_scaled_df
#scaled_df=data[scaled_features]
sel.fit(non_scaled_df/non_scaled_df.mean())

# Create a boolean mask
mask = sel.get_support()
# Apply the mask to create a reduced dataframe
reduced_df_var = non_scaled_df.loc[:, mask]

print("Dimensionality reduced from {} to {}.".format(non_scaled_df.shape[1], reduced_df_var.shape[1]))
```

```python
corr = reduced_df_var.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
# Draw the heatmap
fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corr, mask=mask,  cmap="coolwarm", center=0, linewidths=1, annot=True, fmt=".2f")
plt.savefig('corr_heatmap.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate the correlation matrix and take the absolute value
corr_matrix = reduced_df_var.corr().abs()
# Create a True/False mask and apply it
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)
# List column names of highly correlated features (r > 0.95)
to_drop = [c for c in tri_df.columns if any(tri_df[c] >  0.95)]

# Drop the features in the to_drop list
reduced_df_var_cor = reduced_df_var.drop(to_drop, axis=1)

print("The reduced dataframe has {} columns.".format(reduced_df_var_cor.shape[1]))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Pipeline a scaler and pca selecting 10 components
pipe = Pipeline([('scaler', StandardScaler()),
        		 ('reducer', PCA(n_components=10))])

# Fit the pipe to the data
pipe.fit(reduced_df_var_cor)

# Plot the explained variance ratio
plt.rcParams.update({'font.size': 16})
plt.subplots(figsize=(20,15))
plt.plot(pipe.steps[1][1].explained_variance_ratio_)

plt.xlabel('Principal component index')
plt.ylabel('Explained variance ratio')

plt.savefig('PCA_index.png',dpi=300, bbox_inches='tight')
```

```python
pipe = Pipeline([('scaler', StandardScaler()),
                 ('reducer', PCA(n_components=2))])

# Fit the pipeline to poke_df and transform the data
pc = pipe.fit_transform(reduced_df_var_cor)

# Add the 2 components to poke_cat_df
reduced_df_var_cor['PC 1'] = pc[:, 0]
reduced_df_var_cor['PC 2'] = pc[:, 1]

print(reduced_df_var_cor.head())
```

```python
pca=['PC 1', 'PC 2']
data_final=reduced_df_var_cor[pca]
print(data_final.head())
```
